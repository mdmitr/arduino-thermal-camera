# Detecting edges 

Algorithm for detecting edges. These detected edges will be transformed to points for scanning temperature by my themrmal camera

```python
%pylab inline
from pylab import rcParams
rcParams['figure.figsize'] = 15, 15
```

```python
import matplotlib.pyplot as plt
from scipy import ndimage as ndi
from skimage.color import rgb2gray

import edge_detection as ed
```

## Setting up filename here

```python
filename = "pic1.jpg"
im = ndi.imread(filename)
img = rgb2gray(im)
```

## Show original image

```python
fig = plt.figure()
ax0 = fig.add_subplot(111)
ax0.imshow(im)
ax0.axis('off')
ax0.set_title('Original image', fontsize=12)
plt.show()
```

## Find dark regions

```python
img_dark = ed.find_dark_regions(img, threshold=0.3)
fig = plt.figure()
ax0 = fig.add_subplot(111)
ax0.imshow(img_dark, cmap=plt.cm.gray)
ax0.axis('off')
ax0.set_title('Dark regions', fontsize=12)
plt.show()
```

## Find edges

```python
img_edges = ed.find_edges(img)
fig = plt.figure()
ax0 = fig.add_subplot(111)
ax0.imshow(img_edges, cmap=plt.cm.gray)
ax0.axis('off')
ax0.set_title('Edges', fontsize=12)
plt.show()
```

## Combine dark regions and edges together

```python
img = img_dark + img_edges
fig = plt.figure()
ax0 = fig.add_subplot(111)
ax0.imshow(img, cmap=plt.cm.gray)
ax0.axis('off')
ax0.set_title('Combined image', fontsize=12)
plt.show()
```
